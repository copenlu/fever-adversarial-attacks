```python
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
dataset = []
with open('data/fever2-fixers-dev.jsonl') as out:
    for line in out:
        dataset.append(json.loads(line))
print(len(dataset))
dataset[0]
```

```python
df = json_normalize(dataset)
df.sample(10)
```

```python
df['label'].value_counts(dropna=False, ascending=True)
```

```python
df['transformation'].value_counts(dropna=False, ascending=True)
```

```python
df['attack'].value_counts(dropna=False, ascending=False).to_frame().to_dict()
```

```python
df['annotation'].value_counts(dropna=False, ascending=False)
```
